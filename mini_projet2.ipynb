## Text Pre-processing

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
## encoding the Category (target) 'ham' : 0 and 'spam' : 1
df['spam'] = df['Category'].map( {'spam': 1, 'ham': 0} ).astype(int)
df.head(15)
```

#  the pre-processing of data consist of 4 steps:
* Convert bag of words to lowercase.
* lemmatization of words using NLTK
* Remove punctuation from sentences.
* split sentences into words and remove stopwords. 
* remove digits

```python
def text_processing(mess):

    # Convert bag of words to lowercase
    lower1 = []
    for i in mess:
        lower1.append(i.lower())
    lower1 = ''.join(lower1)    

    # lemmatization of words using NLTK
    doc = nlp(lower1)
    lower2 = []
    for token in doc:
        if(token.lemma_ == '-PRON-'):
            lower2.append(token.string)
        else:
            lower2.append(token.lemma_)
    lower2 = ' '.join(lower2)   
    
    # Remove punctuation from sentences
    nopunc = [char for char in lower2 if char not in string.punctuation]
    
    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)

    # split the data and remove stopwords
    no_stop_words = [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
    final_list = list(np.copy(no_stop_words))
    
    # Remove digits
#     for item in no_stop_words:
#         if item.isalpha() == False:
#             final_list.remove(item)
#         elif len(item) == 1:
#             final_list.remove(item)

    return final_list
```

## Vectorization and create filtering Model using Sklearn

```python
## we are going to calculate the prior probability
from collections import Counter, defaultdict
def occurrences(label_test):
    no_of_examples = len(label_train)
    print(no_of_examples)
    print(Counter(label_test))
    prob = dict(Counter(label_test))
    
    for key in prob.keys():
        prob[key] = prob[key] / float(no_of_examples)
    return prob
prior_dict = occurrences(label_train)
```

#### create a list of unique words in our training data

```python
# create a list of unique words from training set
def list_words_training_set(msg_train):
    list_words = []
    for sentence in msg_train:
        for w in sentence:
            if w not in list_words:
                list_words.append(w)
    return list_words
```

```python
list_words = list_words_training_set(msg_train)
```

#### The bag of words representation 

```python
## transform every message to a bag of wors (word : count) ,a list for Spam and another for Ham message
def vectorizer_couts(msg_train,list_words):
    counts_spam = []
    counts_ham  = []
    for i, message in enumerate(msg_train):
        l = {}
        for j, w in enumerate(list_words):
            l[w] = message.count(w)
        if(label_list_train[i] == 1):
            counts_spam.append(l)
        else:
            counts_ham.append(l)
    return counts_spam,counts_ham
```

```python
message_couts_spam, message_couts_ham = vectorizer_couts(msg_train,list_words)
```

#### liklihood Probabilities P(wordi/spam)

```python
## we calculate likelihood for class Spam 
## create a list of likelihood for every word in our vocabulary
## P(Word|Spam)
def liklihood_W_Spam(message_couts_spam,list_words):
    Prob_w_spam = {}
    for w in list_words:
        nb = 0
        for message_spam in message_couts_spam:
            nb = nb + message_spam[w]
        Prob_w_spam[w] = ((nb + 1)/(counts_words_in_spam_msg + len(list_words)))
    return Prob_w_spam
```

```python
## we calculate likelihood for class Ham 
## create a list of likelihood for every word in our vocabulary
## P(Word|Ham)
def liklihood_W_Ham(message_couts_ham,list_words):
    Prob_w_ham = {}
    for w in list_words:
        nb = 0
        for message_ham in message_couts_ham:
            nb = nb + message_ham[w]
        Prob_w_ham[w] = ((nb + 1)/(counts_words_in_ham_msg + len(list_words)))
    return Prob_w_ham
```

```python
Prob_w_spam = liklihood_W_Spam(message_couts_spam,list_words)
Prob_w_ham  = liklihood_W_Ham(message_couts_ham,list_words)
```

## Test the model

##### To test our model of Naive Bayes we use the liklihood we just calculated and Prior probabilities to calculate Posterior probabilities for every message in our test data:

#### Bayes' theorem


$$ P(Spam \mid Message) = \frac{P(Message \mid Spam) \, P(Spam)}{P(Message)} $$

$$ P(Ham \mid Message) = \frac{P(Message \mid Ham) \, P(Ham)}{P(Message)} $$

$$ P(Ham \mid Message) = \frac{P(Message \mid Ham) \, P(Ham)}{P(Message \mid Ham)*P(Ham) +  P(Message \mid Spam)*P(Spam)}$$


we need to comute one of them for example $ P(Spam \mid Message)$ and conclude the other :
                    
$$ P(Ham \mid Message)  = 1 -  P(Spam \mid Message)$$

So to calassify a message as Spam , this condition should be satidfied:

$$ P(Spam \mid Message) >= 0.5$$


```python
len(msg_test)
```

```python
def evaluate(msg_test):
    result = []
    for msg in msg_test:
        pr_mssg_Spam = 1
        pr_mssg_Ham  = 1
        for word in msg:
            if(word in Prob_w_spam.keys()):
                ## calculate P(Message|Spam)
                pr_mssg_Spam = pr_mssg_Spam*Prob_w_spam[word]
            if(word in Prob_w_ham.keys()):
                ## calculate P(Message|Ham)
                pr_mssg_Ham = pr_mssg_Ham*Prob_w_ham[word]
        ## calculate Posterior probabilities P(Spam|Message)
        a = (pr_mssg_Spam*prior_dict[1])/((pr_mssg_Spam*prior_dict[1])+(pr_mssg_Ham*prior_dict[0]))
        if(a >= 0.5):
            result.append(1)
        else:
            result.append(0)
    return result
    
```

```python
result_prediction = evaluate(msg_test)
```
